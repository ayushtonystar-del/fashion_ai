```python
import torch

from diffusers import StableDiffusionXLPipeline

from PIL import Image

import matplotlib.pyplot as plt

import os
```

```python
torch.backends.cuda.matmul.allow_tf32=True

torch.backends.cudnn.benchmark=True
```

```python
from huggingface_hub import login

login()
```

```python
model_id="stabilityai/stable-diffusion-xl-base-1.0"

pipe=StableDiffusionXLPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    use_safetensors=True
)

pipe=pipe.to(device)
```

```python
pipe.enable_attention_slicing()
```

```python
prompt="""
A luxury fashion model

wearing an elegant white designer dress

editorial photography

highly detailed

8k

realistic

professional lighting

fashion magazine

masterpiece
"""
```

```python
negative_prompt="""
blurry

low quality

deformed

extra fingers

cropped

duplicate

bad anatomy

text

logo

watermark
"""
```

```python
image=pipe(

prompt=prompt,

negative_prompt=negative_prompt,

num_inference_steps=30,

guidance_scale=7.5

).images[0]
```

```python
os.makedirs("outputs/generated_images",exist_ok=True)

image.save("outputs/generated_images/fashion1.png")

print("Saved")
```

```python
prompts=[

"Luxury red evening gown",

"Modern streetwear fashion",

"Casual denim outfit",

"Black leather jacket fashion"
]

images=[]

for p in prompts:

    img=pipe(

        prompt=p,

        negative_prompt=negative_prompt,

        num_inference_steps=30

    ).images[0]

    images.append(img)
```

```python
fig,axs=plt.subplots(2,2,figsize=(10,10))

for ax,img,title in zip(axs.flatten(),images,prompts):

    ax.imshow(img)

    ax.set_title(title)

    ax.axis("off")

plt.show()
```

```python
for i,img in enumerate(images):

    img.save(f"outputs/generated_images/image_{i}.png")

print("All images saved")
```
